```python
import numpy as np
import nconpp as ncp
```

```python
containerList = [
    (np.random.randn(60)+1j*np.random.randn(60)).reshape(3,4,5),
    (np.random.randn(3780)+1j*np.random.randn(3780)).reshape(5, 3, 6, 7, 6),
    (np.random.randn(14)+1j*np.random.randn(14)).reshape(7,2),
    (np.random.randn(8)+1j*np.random.randn(8)).reshape(8),
    (np.random.randn(72)+1j*np.random.randn(72)).reshape(8,9),
    (np.random.randn(81)+1j*np.random.randn(81)).reshape(9,9)
]

legsList = [
    [3, -2, 2],
    [2, 3, 1, 4, 1],
    [4, -1],
    [5],
    [5, -3],
    [6, 6]
]
```

```python
ncp.contract(containerList, legsList)
```

### Function readme_example1

```python
print(xt.readme_example1.__doc__)
```

```python
PI = np.pi
a = np.linspace(0, PI, 100)
b = xt.readme_example1(a)
display(a)
display(np.sin(a).sum())
display(b)
```

### Function vectorize_example1

```python
print(xt.vectorize_example1.__doc__)
```

```python
PI = np.pi
a = np.linspace(0, PI, 10)
b = np.linspace(0, 2*PI, 10)
c = xt.vectorize_example1(a, b)
d = np.sin(a)+np.cos(b)
display(a)
display(b)
display(c)
display(np.isclose(c, d))
```
