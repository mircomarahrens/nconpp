# General k-body hamiltonian

$$
H^{(n)} = \sum_{i=1}^N J_i\mathbf{Z}_i + \sum_{i=1}^N\sum_{j>i} J_{ij}\mathbf{Z}_{i}\mathbf{Z}_{j} + \sum_{i=1}^N\sum_{j>i}\sum_{k>j} J_{ijk}\mathbf{Z}_{i}\mathbf{Z}_{j}\mathbf{Z}_{k} + \ldots
$$
