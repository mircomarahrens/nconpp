```python
from latticegraph import PyLatticeGraph

honeycomb_obc = PyLatticeGraph()

honeycomb_obc.plot_graph()
```

```python
# 1D chain, 4 sites, OBC
chain_obc = PyLatticeGraph(kind="chain")

chain_obc.plot_graph()
```

```python
# Square lattice, 4x4, open boundary conditions
square_obc = PyLatticeGraph(
    kind="square",
    sites=(4, 4),
    unitcell=[((0, +1), (+1, 0), (0, -1), (-1, 0))],
    bcs=("obc", "obc")
)

square_obc.plot_graph()
```

```python
# Honeycomb lattice, 4x4, open boundary conditions
honeycomb_obc = PyLatticeGraph(
    kind="honeycomb",
    sites=(4, 4),
    unitcell=[
        ((0, -1), (0, +1), (+1, 0)),   # sublattice A (even parity sites)
        ((-1, 0), (0, -1), (0, +1))    # sublattice B (odd parity sites)
    ],
    bcs=("obc", "obc")
)

honeycomb_obc.plot_graph()
```

```python
# Triangular lattice, 4x4, open boundary conditions
triangle_obc = PyLatticeGraph(
    kind="triangle",
    sites=(4, 4),
    unitcell=[((0, +1), (+1, 0), (+1, +1), (0, -1), (-1, 0), (-1, -1))],
    bcs=("obc", "obc")
)

triangle_obc.plot_graph()
```
