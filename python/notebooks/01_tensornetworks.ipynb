```python
from graphviz import Digraph
import math

def draw_tensors(node_legs,
                                     #node_width=1,
                                     #edge_len=0.2,
                                     x_spacing=1.0,
                                     per_node_angles=None):
    """
    node_legs: {node: [leg_name, ...]}
    per_node_angles: optional {node: [deg,...]} to control leg angles. If None, legs are evenly spaced.
    Places all main nodes in a horizontal line.
    """
    mpd = Digraph(
        graph_attr={
            "rankdir": "LR",
            "splines": "true",
            "color": "transparent",
            "ranksep": "1",
        },
        node_attr={
            "style": "filled",
            "color": "white",
            "shape": "circle",
            "height": "0.1",
            "fillcolor": "lightgray",
            "fontsize": "10",
        },
        edge_attr={
            "arrowsize": "0.5",
            "fontsize": "8",
        },
        engine="neato"  # allows fixed positions
    )

    nodes = list(node_legs.keys())

    # Place nodes along a horizontal line
    for i, node in enumerate(nodes):
        x = i * x_spacing
        y = 0.0
        mpd.node(str(node),
                 label=str(node),
                 #fixedsize="true",
                 pos=f"{x},{y}!"
                 )

    # Add dangling edge stubs and labels
    for i, node in enumerate(nodes):
        x = i * x_spacing
        y = 0.0

        legs = node_legs.get(node, [])
        k = len(legs)
        if k == 0:
            continue

        if per_node_angles and node in per_node_angles:
            angs = [math.radians(a) for a in per_node_angles[node]]
        else:
            # Evenly spaced around the node
            angs = [math.radians(90 - j * 360.0 / k) for j in range(k)]

        for j, (leg_name, ang) in enumerate(zip(legs, angs)):
            #x1 = x + (node_width + edge_len) * math.cos(ang)
            #y1 = y + (node_width + edge_len) * math.sin(ang)
            
            x1 = x + 0.3 * math.cos(ang)
            y1 = y + 0.3 * math.sin(ang)

            stub_id = f"{node}_leg_{j}"
            mpd.node(stub_id,
                     #label="",
                     label=str(leg_name),
                     shape="point",
                     width="0.02",
                     height="0.02",
                     fixedsize="true",
                     pos=f"{x1},{y1}!")

            mpd.edge(str(node), stub_id, dir="none")

            # label node for the leg name
            label_id = f"{stub_id}_lbl"
            label_x = x1 + 0.2 * math.cos(ang)
            label_y = y1 + 0.2 * math.sin(ang)
            mpd.node(label_id,
                     label=str(leg_name),
                     #shape="point",
                     #shape="plaintext",
                     pos=f"{label_x},{label_y}!",
                     fixedsize="true"
                     )

    return mpd


# Example usage
node_legs = {0: [3, -2, 2], 1: [2, 3, 1, 4, 1], 2: [4, -1], 3: [5], 4: [5, -3], 5: [6, 6]}
g = draw_tensors(node_legs)
# g.render("dangling_graph_line", format="svg", cleanup=True)
# In Jupyter: g

g
```

```python
%matplotlib widget
from ipywidgets import interact
import matplotlib.pyplot as plt
import numpy as np

from nconpp import LatticeGraph

# lg = LatticeGraph("honeycomb", [4,4], [((0, -1), (0, +1), (+1, 0)), ((-1, 0), (0, -1), (0, +1))], ("pbc", "pbc"))


# plot
@interact(lx=4, ly=4, boundary=True)
def plot_lattice(lx, ly, boundary):
    lg = LatticeGraph("honeycomb", [lx, ly])

    _, ax = plt.figure(), plt.axes()

    # vertices
    for i in lg.vertices:
        v = lg.vertices[i]
        if v.boundary:
            ax.scatter(v.coordinate[0], v.coordinate[1], c="r", alpha=0.5)
        else:
            ax.scatter(v.coordinate[0], v.coordinate[1], c="b", alpha=0.5)
        ax.annotate(i, v.coordinate)

    # edges
    for i in lg.edges:
        e = lg.edges[i]
        e_src = lg.vertices[e.src].coordinate
        e_dest = lg.vertices[e.dest].coordinate
        x, y = np.array([e_src, e_dest]).T
        if e.boundary and boundary:
            ax.plot(x, y, c="forestgreen", linestyle=":", alpha=0.8)
            ax.annotate(
                i,
                xy=(np.mean(x) - 0.1, np.mean(y) - 0.1),
                ha="center",
                va="center",
                color="forestgreen",
                fontsize=6,
            )
        else:
            ax.plot(x, y, c="lightsteelblue", linestyle="-", alpha=0.8)
            ax.annotate(
                i,
                xy=(np.mean(x) + 0.1, np.mean(y) + 0.1),
                ha="center",
                va="center",
                color="lightsteelblue",
                fontsize=6,
            )
    plt.show()

    print("e_id\tsrc\tdest\tboundary")
    for i in lg.edges:
        e = lg.edges[i]
        e_src = lg.vertices[e.src].coordinate
        e_dest = lg.vertices[e.dest].coordinate
        print(f'{i}\t{e_src}\t{e_dest}\t{e.boundary}')
```

```python
#from matplotlib import pyplot as plt
from functools import partial
from ipywidgets import interact

import numpy as np

def f(x): return 3*x**2 + 2*x + 1

def plot_function(f, title=None, min=-2.1, max=2.1, color='r', ylim=None):
    x = np.linspace(min,max, 100)[:,None]
    if ylim: 
        plt.ylim(ylim)
    plt.plot(x, f(x), color)
    if title is not None: 
        plt.title(title)

def quad(a,b,c,x):
    return a*x**2 + b*x + c

def mk_quad(a,b,c): return partial(quad, a,b,c)
def noise(x, scale): return np.random.normal(scale=scale, size=x.shape)
def add_noise(x, mult, add): return x * (1+noise(x,mult)) + noise(x,add)

np.random.seed(42)

x = np.linspace(-2, 2)[:,None]
y = add_noise(f(x), 0.15, 1.5)

@interact(a=1.5, b = 1.5, c =1.5)
def plot_quad(a,b,c):
    plt.scatter(x,y)
    plot_function(mk_quad(a,b,c), ylim=(-3, 12))
```
